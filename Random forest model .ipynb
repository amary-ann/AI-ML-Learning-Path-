# Machine Learning model using Linear Regression

```python
#define the input data(x) and output data/prediction(y)
y = housing['median_house_value']

#drop some columns like price as it is in y already
x = housing.drop(['median_house_value', 'households','population','ocean_proximity'], axis = 1)
```

```python
#train data on a linear regression model
#create a linear model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

##### Split the data into the training and testing set

```python
#split the data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 2)
```

##### Check the shape of the training data and testing data 

```python
print((x_train.shape[0] / x.shape[0]) * 100)
print((x_test.shape[0] / x.shape[0]) * 100)
print((y_train.shape[0] / y.shape[0]) * 100)
print((y_test.shape[0] / y.shape[0]) * 100)
```

```python
#train the data 
#call the fit method on the model and pass the training data to it 
model.fit(x_train, y_train)
```

```python
#test and score the model against the test data
model.score(x_test, y_test)
```

# Using Random Forest model 

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

forest_model = RandomForestRegressor(random_state = 1)
forest_model.fit(x_train, y_train)
pred = forest_model.predict(x_test)
mean_absolute_error(y_test, pred)
```

```python
forest_model.score(x_test,y_test)
```
